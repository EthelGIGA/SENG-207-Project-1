```python
data_mode= dataset_1.median()
dataset_2 = dataset_1.fillna(data_mode)
dataset_2
```

### *DATA VISUALIZATION*

### Data Visualization for 'Fixed Acidity'

```python
dat_mean = dataset_2['fixed acidity'].mean()

# Count values Greater than or Leass than Mean
above_mean_count = dataset_2[dataset_2['fixed acidity'] > dat_mean]['fixed acidity'].count()
below_mean_count = dataset_2[dataset_2['fixed acidity'] < dat_mean]['fixed acidity'].count()

# Create pie chart
labels = ['Greater than mean', 'Less than mean']
sizes = [above_mean_count, below_mean_count]
colors = ['#5751E3', '#D30F77']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Fixed Acidity Greater than or Leass than Mean')
plt.show()
```

```python
plot1 = sns.histplot(data=dataset_2, x='fixed acidity',color='#70E9FF')

plot1.set_title("Fixed Acidity Histogram") #title
```
